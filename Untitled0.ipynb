### Extracting Information from Markdown with Python

We can use Python's `re` module (regular expressions) to identify and extract specific patterns from the text. Here's a general approach:

1.  **Read the content:** Make sure the document content is loaded into a string variable.
2.  **Define patterns:** Create regular expressions for the information you want to extract (e.g., headings, bullet points, specific keywords).
3.  **Search and extract:** Use `re.findall()` or `re.search()` to find all occurrences of the patterns.

# **PRICE**

```python
from vnstock import Quote
quote = Quote(symbol= user_input, source='KBS')
```

```python
df = quote.history(length='90', interval='d')
```

# **VNINDEX**

```python
df = quote.history(symbol='VNINDEX', start='2026-06-01', end='2026-06-12', interval='d')
df
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Convert 'time' column to datetime objects if not already converted
df['time'] = pd.to_datetime(df['time'])

# Calculate a 20-day Simple Moving Average (SMA)
df['SMA_20'] = df['close'].rolling(window=20).mean()

# Create the line plot for the closing price of VNINDEX and the SMA
plt.figure(figsize=(12, 6))
sns.lineplot(x='time', y='close', data=df, label='Closing Price')
sns.lineplot(x='time', y='SMA_20', data=df, label='20-day SMA', linestyle='--')
plt.title('VNINDEX Closing Price with 20-day Moving Average')
plt.xlabel('Date')
plt.ylabel('Price')
plt.grid(True, linestyle='--', alpha=0.7)
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```
