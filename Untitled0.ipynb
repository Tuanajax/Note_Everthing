## **Ratio**

```python
finance = Finance(symbol=user_input, source='KBS')
cash_flow_statement = finance.cash_flow(period='year')

# Ensure item_id column is string type and strip whitespace for robust comparison
cash_flow_statement['item_id'] = cash_flow_statement['item_id'].astype(str).str.strip()

# Use .str.contains for robust matching of 'operating_cash_flow'
op_cash_flow = cash_flow_statement[cash_flow_statement['item_id'].str.contains('operating_cash_flow', case=False, na=False)]

# Ensure capex is also found robustly.
capex = cash_flow_statement[cash_flow_statement['item_id'].str.contains('purchase_of_fixed_assets', case=False, na=False)]

print(f"\nFree Cash Flow (FCF) for {user_input} (in VND):\n")

if not op_cash_flow.empty and not capex.empty:
    # Ensure only one row is selected for calculation if multiple exist
    if len(op_cash_flow) > 1:
        print(f"Warning: Multiple entries found for 'operating_cash_flow'. Using the first one.")
        op_cash_flow = op_cash_flow.iloc[[0]]

    if len(capex) > 1:
        print(f"Warning: Multiple entries found for 'purchase_of_fixed_assets'. Using the first one.")
        capex = capex.iloc[[0]]

    year_columns = [col for col in cash_flow_statement.columns if '-Năm' in col]

    fcf_data = {}
    for year_col in year_columns:
        # Check if the year_col exists in the filtered DataFrame before accessing iloc[0]
        op_cf_val = op_cash_flow[year_col].iloc[0] if year_col in op_cash_flow.columns and not op_cash_flow.empty else float('nan')
        capex_val = capex[year_col].iloc[0] if year_col in capex.columns and not capex.empty else float('nan')

        # FCF = Operating Cash Flow + Capital Expenditures (CapEx is usually negative in statement, so we add it as an outflow)
        fcf = op_cf_val + capex_val if pd.notna(op_cf_val) and pd.notna(capex_val) else float('nan')
        fcf_data[year_col.replace('-Năm', '')] = fcf

    for year, value in fcf_data.items():
        if pd.notna(value):
            print(f"  {year}: {value:,.0f}")
        else:
            print(f"  {year}: N/A")
else:
    print("Could not retrieve sufficient data to calculate Free Cash Flow.")
    if op_cash_flow.empty:
        print(f"  'operating_cash_flow' data not found.")
    if capex.empty:
        print(f"  'purchase_of_fixed_assets' data not found.")

print("\n*   **Dòng tiền tự do (Free Cash Flow - FCF):** Là lượng tiền mặt còn lại sau khi một công ty đã thanh toán các chi phí hoạt động và chi phí vốn (CapEx). FCF là một chỉ số quan trọng về khả năng tạo ra tiền mặt thực sự của doanh nghiệp và khả năng chi trả cổ tức, trả nợ, hoặc tái đầu tư.")
```

## **Định giá CP bằng RI**

```python
import pandas as pd
import numpy as np

# --- 1. Define the Residual Income Valuation Function ---
def ri_valuation_tcb(symbol, finance_obj, company_obj, cost_of_equity, growth_rate_ri_phase1=0.08, num_growth_years=5, growth_rate_ri_terminal=0.02):
    print(f"\n🚀 Bắt đầu định giá RI cho {symbol}...")
    print(f"  Chi phí vốn chủ sở hữu (Ke): {cost_of_equity*100:.2f}%")
    print(f"  Tốc độ tăng trưởng RI giai đoạn 1: {growth_rate_ri_phase1*100:.2f}%")
    print(f"  Số năm tăng trưởng giai đoạn 1: {num_growth_years} năm")
    print(f"  Tốc độ tăng trưởng RI vĩnh cửu: {growth_rate_ri_terminal*100:.2f}%")

    # Fetch financial ratios
    try:
        ratios_df = finance_obj.ratio(period='year')
        ratios_df['item_id'] = ratios_df['item_id'].astype(str).str.strip()
    except Exception as e:
        print(f"Error fetching financial ratios for {symbol}: {e}")
        return None

    # Get the latest BVPS
    current_bvps = None
    bvps_item_ids_to_check = ['book_value_per_share_bvps', 'book_value_per_share', 'bvps']
    year_cols = [col for col in ratios_df.columns if '-Năm' in col]

    for bvps_item_id in bvps_item_ids_to_check:
        if bvps_item_id in ratios_df['item_id'].values:
            bvps_data_row = ratios_df[ratios_df['item_id'] == bvps_item_id]
            if not bvps_data_row.empty:
                for year_col in sorted(year_cols, reverse=True):
                    if year_col in bvps_data_row.columns and pd.notna(bvps_data_row[year_col].iloc[0]):
                        current_bvps = bvps_data_row[year_col].iloc[0]
                        break
            if current_bvps is not None:
                break

    # Get the latest EPS
    current_eps = None
    eps_item_ids_to_check = ['earning_per_share_vnd', 'earning_per_share', 'eps', 'trailing_eps']
    for eps_item_id in eps_item_ids_to_check:
        if eps_item_id in ratios_df['item_id'].values:
            eps_data_row = ratios_df[ratios_df['item_id'] == eps_item_id]
            if not eps_data_row.empty:
                for year_col in sorted(year_cols, reverse=True):
                    if year_col in eps_data_row.columns and pd.notna(eps_data_row[year_col].iloc[0]):
                        current_eps = eps_data_row[year_col].iloc[0]
                        break
            if current_eps is not None:
                break

    if current_bvps is None or current_eps is None:
        raise ValueError(f"Không thể lấy đủ dữ liệu BVPS hoặc EPS cho {symbol}. BVPS: {current_bvps}, EPS: {current_eps}")

    print(f"\n  + BVPS hiện tại: {current_bvps:,.2f} VND")
    print(f"  + EPS hiện tại: {current_eps:,.2f} VND")

    # Calculate initial Residual Income (RI0 for next period, based on current EPS and BVPS)
    # A more precise model would project next year's EPS. For simplicity, we use current EPS as basis.
    ri0 = current_eps - (cost_of_equity * current_bvps)
    print(f"  + Lợi nhuận thặng dư hiện tại (RI_0): {ri0:,.2f} VND")

    projected_ris = {}
    total_pv_ri = 0

    # Project Residual Income for Phase 1 (high growth)
    print("\n  Dự phóng Lợi nhuận Thặng dư (Residual Income):")
    for i in range(1, num_growth_years + 1):
        if i == 1:
            ri_current = ri0 * (1 + growth_rate_ri_phase1)
        else:
            ri_current = projected_ris[i - 1] * (1 + growth_rate_ri_phase1)

        discount_factor = 1 / ((1 + cost_of_equity) ** i)
        pv_ri = ri_current * discount_factor
        projected_ris[i] = ri_current
        total_pv_ri += pv_ri
        print(f"    RI năm {i}: {ri_current:,.2f} VND (PV: {pv_ri:,.2f} VND)")

    # Calculate Terminal Value of Residual Income (after num_growth_years)
    last_projected_ri = projected_ris[num_growth_years]
    terminal_value_ri = (last_projected_ri * (1 + growth_rate_ri_terminal)) / (cost_of_equity - growth_rate_ri_terminal)

    # Discount Terminal Value of RI back to present
    pv_terminal_value_ri = terminal_value_ri / ((1 + cost_of_equity) ** num_growth_years)

    print(f"\n  Lợi nhuận thặng dư (RI) tại năm cuối cùng ({num_growth_years}): {last_projected_ri:,.2f} VND")
    print(f"  Giá trị vĩnh cửu của RI tại năm {num_growth_years}: {terminal_value_ri:,.2f} VND")
    print(f"  Giá trị hiện tại của Giá trị vĩnh cửu RI: {pv_terminal_value_ri:,.2f} VND")

    # Intrinsic Value = Current BVPS + PV of all projected RIs + PV of Terminal Value of RI
    intrinsic_value_per_share = current_bvps + total_pv_ri + pv_terminal_value_ri

    print("\n" + "="*50)
    print(f"📊 KẾT QUẢ ĐỊNH GIÁ RI CHO {symbol}")
    print("="*50)
    print(f"  BVPS hiện tại: {current_bvps:,.2f} VND")
    print(f"  Tổng giá trị hiện tại của RI dự phóng: {total_pv_ri:,.2f} VND")
    print(f"  Giá trị hiện tại của Giá trị vĩnh cửu RI: {pv_terminal_value_ri:,.2f} VND")
    print("-"*50)
    print(f"  🎯 GIÁ TRỊ NỘI TẠI MỖI CỔ PHIẲU: {intrinsic_value_per_share:,.2f} VND")
    print("="*50)

    return intrinsic_value_per_share

# --- 2. Set Assumptions for RI Model ---
# We use the 'finance' and 'company' objects initialized earlier for user_input
# These values can be adjusted based on detailed analysis of the company and market.
default_ke = 0.12 # Cost of Equity (Ke) - Example: 12%
growth_rate_ri_phase1 = 0.08 # Residual Income growth rate for the first phase (e.g., 8%)
num_ri_growth_years = 5 # Number of years for the initial growth phase
growth_rate_ri_terminal = 0.02 # Terminal growth rate for Residual Income (e.g., 2%)

# --- 3. Execute the RI Valuation ---
try:
    intrinsic_value_ri = ri_valuation_tcb(
        symbol=user_input,
        finance_obj=finance,
        company_obj=company,
        cost_of_equity=default_ke,
        growth_rate_ri_phase1=growth_rate_ri_phase1,
        num_growth_years=num_ri_growth_years,
        growth_rate_ri_terminal=growth_rate_ri_terminal
    )
    print(f"\nGiá trị nội tại của {user_input} theo mô hình RI là: {intrinsic_value_ri:,.2f} VND/cổ phiếu")
except ValueError as e:
    print(f"Lỗi định giá RI: {e}")
    intrinsic_value_ri = None
except Exception as e:
    print(f"Đã xảy ra lỗi không mong muốn trong quá trình định giá RI: {e}")
    intrinsic_value_ri = None

# Get current market price for comparison
current_market_price = None
if 'df' in locals() and not df.empty:
    current_market_price = df['close'].iloc[-1] # Get the latest closing price from the existing DataFrame
else:
    # Fallback if df is not available or empty, try fetching directly
    try:
        # Fetch price for the last known date in the context (2026-06-12)
        current_market_price_df = quote.history(symbol=user_input, start='2026-06-12', end='2026-06-12', interval='d')
        if not current_market_price_df.empty:
            current_market_price = current_market_price_df['close'].iloc[-1]
        else:
            print("Could not fetch current market price for comparison.")
    except Exception as e:
        print(f"Error fetching current market price: {e}")

# Scale current_market_price if it's too small compared to intrinsic value, assuming it's in thousands.
# A simple heuristic: if market price is < 100 and intrinsic value is > 10000, scale it.
if current_market_price is not None and current_market_price < 100 and intrinsic_value_ri is not None and intrinsic_value_ri > 10000:
    current_market_price_scaled = current_market_price * 1000
    print(f"\nGiá thị trường hiện tại của {user_input} (đã điều chỉnh): {current_market_price_scaled:,.2f} VND/cổ phiếu")
else:
    current_market_price_scaled = current_market_price
    print(f"\nGiá thị trường hiện tại của {user_input}: {current_market_price:,.2f} VND/cổ phiếu")

# Calculate Margin of Safety using the scaled market price
if intrinsic_value_ri is not None and current_market_price_scaled is not None and current_market_price_scaled > 0:
    margin_of_safety = ((intrinsic_value_ri - current_market_price_scaled) / current_market_price_scaled) * 100
    print(f"Biên độ an toàn theo mô hình RI: {margin_of_safety:,.2f}%")
    if margin_of_safety > 0:
        print(f"Điều này cho thấy giá trị nội tại theo RI cao hơn thị giá hiện tại, với biên độ an toàn là {margin_of_safety:,.2f}%.")
    else:
        print(f"Điều này cho thấy giá trị nội tại theo RI thấp hơn hoặc bằng thị giá hiện tại.")
else:
    print("Không thể tính toán biên độ an toàn do thiếu dữ liệu hoặc giá thị trường không hợp lệ.")
```

## Định giá cổ phiếu TCB bằng phương pháp DCF

```python
import pandas as pd
import numpy as np

# --- 1. Get Outstanding Shares ---
# Assuming company.overview() has been called and the result is available in a variable or can be re-called.
# If `company_overview_df` is not directly available, we can re-call company.overview()
company_overview_df = company.overview()

# Ensure 'outstanding_shares' is treated as a numeric value. It might be a string from API.
# Example: '7086240414' needs to be converted to float/int.
outstanding_shares_str = company_overview_df['outstanding_shares'].iloc[0]
outstanding_shares = float(str(outstanding_shares_str).replace(',', ''))

print(f"Số lượng cổ phiếu đang lưu hành của {user_input}: {outstanding_shares:,.0f}")

# --- 2. Define DCF Function ---
def dcf_valuation_tcb(fcf_data_dict, outstanding_shares, growth_rate_phase1=0.07, num_growth_years=3, growth_rate_phase2=0.05, terminal_growth_rate=0.02, wacc=0.10):
    print(f"\n🚀 Bắt đầu định giá DCF cho {user_input}...")
    print(f"  Tốc độ tăng trưởng FCF giai đoạn 1: {growth_rate_phase1*100:.2f}%")
    print(f"  Số năm tăng trưởng giai đoạn 1: {num_growth_years} năm")
    print(f"  Tốc độ tăng trưởng FCF giai đoạn 2: {growth_rate_phase2*100:.2f}%")
    print(f"  Tốc độ tăng trưởng vĩnh cửu: {terminal_growth_rate*100:.2f}%")
    print(f"  Chi phí vốn bình quân (WACC): {wacc*100:.2f}%")

    # Convert fcf_data_dict keys to integers for sorting and projection
    fcf_years = sorted([int(y) for y in fcf_data_dict.keys()])
    latest_fcf_year = fcf_years[-1]
    latest_fcf_value = fcf_data_dict[str(latest_fcf_year)]

    print(f"\n  + FCF năm gần nhất ({latest_fcf_year}): {latest_fcf_value:,.0f} VND")

    projected_fcf = {}
    total_pv_fcf = 0

    # Project FCF for Phase 1 (high growth)
    print("\n  Dự phóng FCF giai đoạn tăng trưởng cao:")
    for i in range(1, num_growth_years + 1):
        projected_year = latest_fcf_year + i
        if i == 1:
            fcf_current = latest_fcf_value * (1 + growth_rate_phase1)
        else:
            fcf_current = projected_fcf[projected_year - 1] * (1 + growth_rate_phase1)

        discount_factor = 1 / ((1 + wacc) ** i)
        pv_fcf = fcf_current * discount_factor
        projected_fcf[projected_year] = fcf_current
        total_pv_fcf += pv_fcf
        print(f"    FCF năm {projected_year}: {fcf_current:,.0f} VND (PV: {pv_fcf:,.0f} VND)")

    # Project FCF for Phase 2 (mid growth, after num_growth_years)
    print("\n  Dự phóng FCF giai đoạn tăng trưởng trung bình:")
    last_fcf_phase1_year = latest_fcf_year + num_growth_years
    last_fcf_phase1_value = projected_fcf[last_fcf_phase1_year]

    num_phase2_years = 5 - num_growth_years # Assuming a total of 5 projection years

    for i in range(1, num_phase2_years + 1):
        projection_index = num_growth_years + i # Overall index for discounting
        projected_year = last_fcf_phase1_year + i
        if i == 1:
            fcf_current = last_fcf_phase1_value * (1 + growth_rate_phase2)
        else:
            fcf_current = projected_fcf[projected_year - 1] * (1 + growth_rate_phase2)

        discount_factor = 1 / ((1 + wacc) ** projection_index)
        pv_fcf = fcf_current * discount_factor
        projected_fcf[projected_year] = fcf_current
        total_pv_fcf += pv_fcf
        print(f"    FCF năm {projected_year}: {fcf_current:,.0f} VND (PV: {pv_fcf:,.0f} VND)")

    # Calculate Terminal Value
    print("\n  Tính toán Giá trị Vĩnh cửu (Terminal Value):")
    final_projected_year = latest_fcf_year + num_growth_years + num_phase2_years
    final_projected_fcf = projected_fcf[final_projected_year]

    terminal_value = (final_projected_fcf * (1 + terminal_growth_rate)) / (wacc - terminal_growth_rate)

    # Discount Terminal Value back to present
    pv_terminal_value = terminal_value / ((1 + wacc) ** (final_projected_year - latest_fcf_year))

    print(f"    FCF tại năm cuối cùng ({final_projected_year}) cho TV: {final_projected_fcf:,.0f} VND")
    print(f"    Giá trị vĩnh cửu: {terminal_value:,.0f} VND")
    print(f"    Giá trị hiện tại của Giá trị vĩnh cửu: {pv_terminal_value:,.0f} VND")

    # Calculate Enterprise Value
    enterprise_value = total_pv_fcf + pv_terminal_value

    # Calculate Intrinsic Value Per Share
    intrinsic_value_per_share = enterprise_value / outstanding_shares

    print("\n" + "="*50)
    print(f"📊 KẾT QUẢ ĐỊNH GIÁ DCF CHO {user_input}")
    print("="*50)
    print(f"  Tổng giá trị hiện tại của FCF dự phóng: {total_pv_fcf:,.0f} VND")
    print(f"  Giá trị hiện tại của Giá trị vĩnh cửu:    {pv_terminal_value:,.0f} VND")
    print(f"  Tổng Giá trị Doanh nghiệp (Enterprise Value): {enterprise_value:,.0f} VND")
    print(f"  Số lượng cổ phiếu đang lưu hành:       {outstanding_shares:,.0f}")
    print("-"*50)
    print(f"  🎯 GIÁ TRỊ NỘI TẠI MỖI CỔ PHIẾU: {intrinsic_value_per_share:,.2f} VND")
    print("="*50)

    return intrinsic_value_per_share, enterprise_value

# --- 3. Call the DCF function with TCB's data ---
# fcf_data is already available from previous cells.
# outstanding_shares is now retrieved.

# Define assumptions
growth_rate_phase1 = 0.10  # 10% for the first 3 years
num_growth_years = 3
growth_rate_phase2 = 0.05  # 5% for the next 2 years (total 5 projection years)
terminal_growth_rate = 0.02 # 2% long-term growth
wacc = 0.12 # 12% Cost of Capital

intrinsic_value, enterprise_value = dcf_valuation_tcb(
    fcf_data,
    outstanding_shares,
    growth_rate_phase1=growth_rate_phase1,
    num_growth_years=num_growth_years,
    growth_rate_phase2=growth_rate_phase2,
    terminal_growth_rate=terminal_growth_rate,
    wacc=wacc
)

print(f"\nGiá trị nội tại của TCB theo DCF là: {intrinsic_value:,.2f} VND/cổ phiếu")
```

## Định giá Cổ phiếu bằng Mô hình Chiết khấu Cổ tức (DDM)

Mô hình Chiết khấu Cổ tức (Dividend Discount Model - DDM) là một phương pháp định giá cổ phiếu dựa trên lý thuyết rằng giá trị nội tại của một cổ phiếu bằng tổng giá trị hiện tại của tất cả các khoản cổ tức trong tương lai mà nó dự kiến sẽ chi trả. Đây là một phương pháp phổ biến cho các công ty đã trưởng thành và có lịch sử chi trả cổ tức ổn định.

### Các giả định chính của DDM:
1.  **Cổ tức là nguồn thu nhập duy nhất:** Mô hình giả định rằng nhà đầu tư chỉ nhận được lợi nhuận từ cổ tức (không tính lợi nhuận từ việc bán cổ phiếu).
2.  **Khả năng dự báo cổ tức:** Cổ tức trong tương lai có thể được ước tính một cách hợp lý.
3.  **Tỷ suất sinh lợi yêu cầu (Cost of Equity - Ke):** Đây là tỷ lệ chiết khấu được sử dụng để đưa các khoản cổ tức tương lai về giá trị hiện tại.

### DDM hai giai đoạn:
Để phản ánh thực tế rằng các công ty thường có các giai đoạn tăng trưởng khác nhau, chúng ta thường sử dụng DDM đa giai đoạn. Ở đây, chúng ta sẽ áp dụng DDM hai giai đoạn:
*   **Giai đoạn 1 (High Growth Phase):** Cổ tức tăng trưởng với tốc độ cao hơn trong một số năm nhất định.
*   **Giai đoạn 2 (Terminal Growth Phase - Gordon Growth Model):** Sau giai đoạn tăng trưởng cao, cổ tức tăng trưởng với tốc độ ổn định (thường là tốc độ tăng trưởng vĩnh cửu) đến vô cùng.

Công thức tính giá trị nội tại theo DDM:

$$ V_0 = \sum_{t=1}^{n} \frac{D_t}{(1+K_e)^t} + \frac{P_n}{(1+K_e)^n} $$

Trong đó:
*   $V_0$: Giá trị nội tại của cổ phiếu hôm nay.
*   $D_t$: Cổ tức dự kiến nhận được vào cuối năm $t$.
*   $K_e$: Chi phí vốn chủ sở hữu (Cost of Equity).
*   $n$: Số năm của giai đoạn tăng trưởng cao.
*   $P_n$: Giá trị cuối kỳ (Terminal Value) của cổ phiếu vào cuối năm $n$, thường được tính bằng Mô hình tăng trưởng Gordon:
    $$ P_n = \frac{D_{n+1}}{(K_e - g)} $$
    *   $D_{n+1}$: Cổ tức dự kiến vào năm sau giai đoạn tăng trưởng cao ($D_n \times (1+g)$).
    *   $g$: Tốc độ tăng trưởng cổ tức vĩnh cửu (terminal growth rate).

```python
import pandas as pd
import numpy as np

# --- 1. Define the Dividend Discount Model (DDM) Valuation Function ---
def ddm_valuation(symbol, finance_obj, cost_of_equity, market_price,
                  growth_rate_phase1=0.07, num_growth_years=5,
                  terminal_growth_rate=0.03):

    print(f"\n🚀 Bắt đầu định giá DDM cho {symbol}...")
    print(f"  Chi phí vốn chủ sở hữu (Ke): {cost_of_equity*100:.2f}%")
    print(f"  Tốc độ tăng trưởng cổ tức giai đoạn 1: {growth_rate_phase1*100:.2f}%")
    print(f"  Số năm tăng trưởng giai đoạn 1: {num_growth_years} năm")
    print(f"  Tốc độ tăng trưởng cổ tức vĩnh cửu: {terminal_growth_rate*100:.2f}%")

    # Fetch financial ratios
    try:
        ratios_df = finance_obj.ratio(period='year')
        ratios_df['item_id'] = ratios_df['item_id'].astype(str).str.strip()
    except Exception as e:
        print(f"Error fetching financial ratios for {symbol}: {e}")
        return None

    # Get the latest Dividend Yield
    dividend_yield = None
    dividend_yield_item_id = 'dividend_yield' # Assuming this is the item_id for Tỷ suất cổ tức
    year_cols = [col for col in ratios_df.columns if '-Năm' in col]

    if dividend_yield_item_id in ratios_df['item_id'].values:
        dy_data_row = ratios_df[ratios_df['item_id'] == dividend_yield_item_id]
        if not dy_data_row.empty:
            for year_col in sorted(year_cols, reverse=True):
                if year_col in dy_data_row.columns and pd.notna(dy_data_row[year_col].iloc[0]):
                    # Assume API returns dividend yield as a percentage (e.g., 1.5 for 1.5%), convert to decimal
                    dividend_yield = dy_data_row[year_col].iloc[0] / 100
                    break

    # Get the latest EPS
    current_eps = None
    eps_item_ids_to_check = ['earning_per_share_vnd', 'earning_per_share', 'eps', 'trailing_eps']
    for eps_item_id in eps_item_ids_to_check:
        if eps_item_id in ratios_df['item_id'].values:
            eps_data_row = ratios_df[ratios_df['item_id'] == eps_item_id]
            if not eps_data_row.empty:
                for year_col in sorted(year_cols, reverse=True):
                    if year_col in eps_data_row.columns and pd.notna(eps_data_row[year_col].iloc[0]):
                        current_eps = eps_data_row[year_col].iloc[0]
                        break
            if current_eps is not None:
                break

    # Calculate Current DPS
    current_dps = None
    assumed_payout_ratio = 0.40 # Default assumed payout ratio for fallback (e.g., 40% for a stable bank)

    # Attempt 1: Calculate DPS from dividend yield and market price
    if dividend_yield is not None and market_price is not None and market_price > 0:
        calculated_dps_from_yield = dividend_yield * market_price
        print(f"  + Tỷ suất cổ tức (mới nhất): {dividend_yield*100:.2f}%")
        print(f"  + Cổ tức hiện tại trên mỗi cổ phiếu (DPS0): {calculated_dps_from_yield:,.2f} VND (ước tính từ giá thị trường và tỷ suất cổ tức)")

        # Heuristic check: if calculated_dps_from_yield is suspiciously low compared to EPS
        if current_eps is not None and current_eps > 0 and calculated_dps_from_yield < (0.05 * current_eps): # If DPS is less than 5% of EPS
            print(f"  --> Cảnh báo: DPS từ tỷ suất cổ tức quá thấp ({calculated_dps_from_yield:,.2f} VND) so với EPS ({current_eps:,.2f} VND).")
            print(f"  --> Sẽ sử dụng EPS ({current_eps:,.2f} VND) và tỷ lệ chi trả cổ tức giả định ({assumed_payout_ratio*100:.0f}%) thay thế.")
            current_dps = current_eps * assumed_payout_ratio
            print(f"  + Cổ tức hiện tại trên mỗi cổ phiếu (DPS0) (điều chỉnh): {current_dps:,.2f} VND (ước tính từ EPS và tỷ lệ chi trả)")
        else:
            current_dps = calculated_dps_from_yield
    elif current_eps is not None:
        # Fallback: If dividend yield not available or problematic, estimate DPS using EPS and an assumed payout ratio
        current_dps = current_eps * assumed_payout_ratio
        print(f"  + EPS hiện tại: {current_eps:,.2f} VND")
        print(f"  + Tỷ lệ chi trả cổ tức giả định: {assumed_payout_ratio*100:.2f}%")
        print(f"  + Cổ tức hiện tại trên mỗi cổ phiếu (DPS0): {current_dps:,.2f} VND (ước tính từ EPS và tỷ lệ chi trả)")
    else:
        raise ValueError("Không thể ước tính Cổ tức trên mỗi cổ phiếu (DPS) do thiếu dữ liệu.")

    if current_dps is None or current_dps <= 0:
        raise ValueError("Cổ tức trên mỗi cổ phiếu (DPS) không hợp lệ hoặc bằng 0.")

    projected_dps = {}
    total_pv_dps = 0

    # Project Dividends for Phase 1 (high growth)
    print("\n  Dự phóng Cổ tức (Dividends) giai đoạn tăng trưởng cao:")
    for i in range(1, num_growth_years + 1):
        if i == 1:
            dps_current = current_dps * (1 + growth_rate_phase1)
        else:
            dps_current = projected_dps[i - 1] * (1 + growth_rate_phase1)

        discount_factor = 1 / ((1 + cost_of_equity) ** i)
        pv_dps = dps_current * discount_factor
        projected_dps[i] = dps_current
        total_pv_dps += pv_dps
        print(f"    DPS năm {i}: {dps_current:,.2f} VND (PV: {pv_dps:,.2f} VND)")

    # Calculate Terminal Value using Gordon Growth Model at the end of Phase 1
    last_projected_dps = projected_dps[num_growth_years]
    terminal_value = (last_projected_dps * (1 + terminal_growth_rate)) / (cost_of_equity - terminal_growth_rate)

    # Discount Terminal Value back to present
    pv_terminal_value = terminal_value / ((1 + cost_of_equity) ** num_growth_years)

    print(f"\n  Cổ tức (DPS) tại năm cuối cùng ({num_growth_years}): {last_projected_dps:,.2f} VND")
    print(f"  Giá trị cuối kỳ (Terminal Value) tại năm {num_growth_years}: {terminal_value:,.2f} VND")
    print(f"  Giá trị hiện tại của Giá trị cuối kỳ: {pv_terminal_value:,.2f} VND")

    # Intrinsic Value = PV of all projected DPS + PV of Terminal Value
    intrinsic_value_per_share = total_pv_dps + pv_terminal_value

    print("\n" + "="*50)
    print(f"📊 KẾT QUẢ ĐỊNH GIÁ DDM CHO {symbol}")
    print("="*50)
    print(f"  Tổng giá trị hiện tại của DPS dự phóng: {total_pv_dps:,.2f} VND")
    print(f"  Giá trị hiện tại của Giá trị cuối kỳ: {pv_terminal_value:,.2f} VND")
    print("-"*50)
    print(f"  🎯 GIÁ TRỊ NỘI TẠI MỖI CỔ PHIẾU: {intrinsic_value_per_share:,.2f} VND")
    print("="*50)

    return intrinsic_value_per_share

# --- 2. Set Assumptions for DDM Model ---
# Use the 'finance' and 'company' objects initialized earlier for user_input (VCB)
# Use the scaled market price for VCB from previous cell: current_market_price_scaled
current_price_vcb_scaled = current_market_price_scaled # From previous cell

# These values can be adjusted based on detailed analysis of the company and market.
default_ke_ddm = 0.12 # Cost of Equity (Ke) - Example: 12%
growth_rate_dps_phase1 = 0.07 # Dividend growth rate for the first phase (e.g., 7%)
num_dps_growth_years = 5 # Number of years for the initial growth phase
terminal_growth_rate_ddm = 0.03 # Terminal growth rate for Dividends (e.g., 3%)

# --- 3. Execute the DDM Valuation ---
try:
    intrinsic_value_ddm = ddm_valuation(
        symbol=user_input,
        finance_obj=finance,
        cost_of_equity=default_ke_ddm,
        market_price=current_price_vcb_scaled,
        growth_rate_phase1=growth_rate_dps_phase1,
        num_growth_years=num_dps_growth_years,
        terminal_growth_rate=terminal_growth_rate_ddm
    )
    print(f"\nGiá trị nội tại của {user_input} theo mô hình DDM là: {intrinsic_value_ddm:,.2f} VND/cổ phiếu")
except ValueError as e:
    print(f"Lỗi định giá DDM: {e}")
    intrinsic_value_ddm = None
except Exception as e:
    print(f"Đã xảy ra lỗi không mong muốn trong quá trình định giá DDM: {e}")
    intrinsic_value_ddm = None
```

## Giải thích mô hình Dự báo Bayesian cho Giá cổ phiếu

Dự báo Bayesian là một phương pháp thống kê cho phép chúng ta cập nhật niềm tin của mình về các tham số mô hình hoặc các dự đoán trong tương lai khi có dữ liệu mới. Không giống như các phương pháp dự báo truyền thống thường cung cấp một ước tính điểm duy nhất, phương pháp Bayesian cung cấp một phân phối xác suất đầy đủ cho các dự đoán, giúp chúng ta định lượng và hiểu rõ hơn về sự không chắc chắn.

### Tại sao sử dụng Bayesian cho dự báo giá cổ phiếu?

1.  **Định lượng sự không chắc chắn:** Thị trường chứng khoán vốn dĩ không chắc chắn. Các mô hình Bayesian cung cấp không chỉ một điểm dự báo mà còn là khoảng tin cậy (credible intervals), cho chúng ta biết phạm vi có thể của giá cổ phiếu trong tương lai và mức độ tin cậy của chúng ta đối với dự báo đó. Điều này rất quan trọng để đánh giá rủi ro.
2.  **Kết hợp kiến thức tiền nghiệm (Prior Knowledge):** Phương pháp Bayesian cho phép chúng ta đưa vào mô hình những niềm tin hoặc thông tin đã biết trước đó (priors) về các tham số. Ví dụ, chúng ta có thể có niềm tin rằng giá cổ phiếu không thể giảm quá một mức nhất định dựa trên các giới hạn biên độ, hoặc một số chỉ số kỹ thuật có ảnh hưởng tích cực/tiêu cực lên giá.
3.  **Cập nhật liên tục:** Mô hình Bayesian có thể dễ dàng được cập nhật với dữ liệu mới. Khi có thêm dữ liệu lịch sử giá hoặc các chỉ số mới, mô hình có thể điều chỉnh phân phối hậu nghiệm (posterior distribution) của nó, dẫn đến các dự báo chính xác hơn.
4.  **Xử lý mô hình phức tạp và dữ liệu nhỏ:** Phương pháp Bayesian rất linh hoạt trong việc mô hình hóa các mối quan hệ phức tạp và có thể hoạt động tốt ngay cả với bộ dữ liệu tương đối nhỏ, bằng cách tận dụng thông tin tiền nghiệm.

### Các bước cơ bản trong dự báo Bayesian:

1.  **Xác định mô hình:** Chọn một mô hình thống kê phù hợp để mô tả mối quan hệ giữa các biến đầu vào (ví dụ: giá quá khứ, khối lượng, chỉ số kỹ thuật) và biến đầu ra (giá cổ phiếu trong tương lai).
2.  **Thiết lập các phân phối tiền nghiệm (Prior Distributions):** Xác định niềm tin ban đầu của bạn về các tham số của mô hình. Nếu không có thông tin cụ thể, có thể sử dụng các prior không thông tin (non-informative priors).
3.  **Xây dựng hàm khả năng (Likelihood Function):** Định nghĩa cách dữ liệu được quan sát phù hợp với mô hình của bạn, đưa ra xác suất của dữ liệu dựa trên các tham số của mô hình.
4.  **Tính toán phân phối hậu nghiệm (Posterior Distribution):** Đây là trọng tâm của suy luận Bayesian. Phân phối hậu nghiệm kết hợp thông tin từ prior và likelihood để đưa ra niềm tin cập nhật về các tham số sau khi xem xét dữ liệu. Thường được thực hiện thông qua các thuật toán Markov Chain Monte Carlo (MCMC).
5.  **Dự báo:** Sử dụng phân phối hậu nghiệm của các tham số để tạo ra các dự đoán về các giá trị tương lai của giá cổ phiếu, bao gồm cả khoảng tin cậy.

### Ví dụ đơn giản: Bayesian Linear Regression

Chúng ta sẽ bắt đầu với một mô hình đơn giản: Hồi quy tuyến tính Bayesian, trong đó giá cổ phiếu trong tương lai được dự báo dựa trên giá cổ phiếu trong quá khứ. Thay vì nhận được một giá trị duy nhất cho hệ số hồi quy, chúng ta sẽ nhận được một phân phối các giá trị có thể, phản ánh sự không chắc chắn.

Chúng ta sẽ sử dụng giá đóng cửa lịch sử của TCB (`df_history`) để xây dựng một mô hình hồi quy tuyến tính đơn giản, dự đoán giá ngày mai dựa trên giá hôm nay.

```python
import pymc as pm
import arviz as az
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Chuẩn bị dữ liệu từ df_history
# Đảm bảo df_history có cột 'close' và đã được sắp xếp theo thời gian
df_bayesian = df[['time', 'close']].copy()
df_bayesian['time'] = pd.to_datetime(df_bayesian['time'])
df_bayesian = df_bayesian.sort_values(by='time').reset_index(drop=True)

# Tạo biến độc lập (X) là giá đóng cửa ngày hôm trước và biến phụ thuộc (y) là giá đóng cửa ngày hôm nay
# Đây là một dạng hồi quy tự hồi quy AR(1) đơn giản
X_bayesian = df_bayesian['close'].iloc[:-1].values
y_bayesian = df_bayesian['close'].iloc[1:].values

# Lấy ngày cuối cùng của dữ liệu để dự đoán ngày tiếp theo
last_close_price = df_bayesian['close'].iloc[-1]
last_date = df_bayesian['time'].iloc[-1]

# Xây dựng mô hình hồi quy tuyến tính Bayesian
print(f"\nBuilding Bayesian Linear Regression model for {user_input}...")
with pm.Model() as linear_model:
    # Priors cho các tham số của mô hình
    # intercept (alpha): hệ số chặn
    # slope (beta): hệ số góc (ảnh hưởng của giá hôm trước lên giá hôm nay)
    # sigma: độ lệch chuẩn của sai số (độ không chắc chắn còn lại)
    alpha = pm.Normal('alpha', mu=0, sigma=10) # prior cho hệ số chặn
    beta = pm.Normal('beta', mu=1, sigma=5)   # prior cho hệ số góc (giả định ban đầu là giá hôm nay gần bằng giá hôm qua)
    sigma = pm.HalfNormal('sigma', sigma=1) # prior cho độ lệch chuẩn của sai số (không thể âm)

    # Hàm khả năng (Likelihood) - giá trị y được tạo ra từ phân phối Normal
    # Giá trị kỳ vọng (mu) được tính từ mô hình hồi quy tuyến tính
    mu = pm.Deterministic('mu', alpha + beta * X_bayesian)
    Y_obs = pm.Normal('Y_obs', mu=mu, sigma=sigma, observed=y_bayesian)

    # Lấy mẫu từ phân phối hậu nghiệm (Posterior Distribution) bằng MCMC (NUTS sampler)
    # draw: số lượng mẫu sau burn-in
    # tune: số lượng mẫu burn-in (loại bỏ để đảm bảo tính hội tụ)
    trace = pm.sample(draws=2000, tune=1000, chains=2, random_seed=42, return_inferencedata=True)

print("\nBayesian model built and sampled successfully!")

# Kiểm tra tóm tắt phân phối hậu nghiệm của các tham số
print("\nSummary of Posterior Distribution:")
az.summary(trace, var_names=['alpha', 'beta', 'sigma'])

# Dự đoán giá cho ngày tiếp theo
# Sử dụng last_close_price như là X_new để dự đoán y_new
with linear_model:
    posterior_predictive = pm.sample_posterior_predictive(trace, var_names=['mu'], extend_inferencedata=True)

# Lấy các mẫu dự đoán cho giá đóng cửa ngày tiếp theo
predicted_next_day_prices = trace.posterior['alpha'].values.flatten() + trace.posterior['beta'].values.flatten() * last_close_price

mean_predicted_price = np.mean(predicted_next_day_prices)
higher_bound = np.percentile(predicted_next_day_prices, 97.5)
lower_bound = np.percentile(predicted_next_day_prices, 2.5)

print(f"\n--- Bayesian Price Forecast for {last_date.date() + pd.Timedelta(days=1)} ---")
print(f"  Giá đóng cửa cuối cùng được biết ({last_date.date()}): {last_close_price:,.2f} VND")
print(f"  Giá dự báo trung bình cho ngày tiếp theo: {mean_predicted_price:,.2f} VND")
print(f"  Khoảng tin cậy 95% (2.5% - 97.5%): ({lower_bound:,.2f} VND - {higher_bound:,.2f} VND)")


# Visualize the posterior predictive distribution
plt.figure(figsize=(10, 6))
sns.histplot(predicted_next_day_prices, kde=True, bins=50, color='skyblue')
plt.axvline(mean_predicted_price, color='red', linestyle='--', label=f'Mean Prediction: {mean_predicted_price:,.2f}')
plt.axvline(lower_bound, color='green', linestyle=':', label=f'95% Credible Interval')
plt.axvline(higher_bound, color='green', linestyle=':')
plt.title(f'Phân phối dự đoán giá TCB vào ngày {last_date.date() + pd.Timedelta(days=1)}')
plt.xlabel('Giá cổ phiếu dự kiến (VND)')
plt.ylabel('Mật độ / Tần suất')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

## Phân tích các chỉ số định giá khác: P/B và ROE

Ngoài P/E và FCF, **P/B (Price-to-Book)** và **ROE (Return on Equity)** là hai chỉ số cực kỳ quan trọng trong định giá cổ phiếu ngân hàng và các tổ chức tài chính khác. Chúng ta sẽ phân tích ý nghĩa của từng chỉ số và so sánh chúng giữa TCB và các ngân hàng khác.

*   **P/B Ratio (Price-to-Book Ratio):**
    *   Chỉ số này so sánh giá trị thị trường của cổ phiếu với giá trị sổ sách của mỗi cổ phiếu. Đối với ngân hàng, giá trị sổ sách (book value) thường phản ánh tốt giá trị tài sản ròng, vì phần lớn tài sản của ngân hàng là tài sản tài chính có thể định giá tương đối chính xác.
    *   P/B < 1 có thể cho thấy cổ phiếu bị định giá thấp so với giá trị tài sản ròng. P/B > 1 cho thấy thị trường đánh giá công ty có khả năng tạo ra lợi nhuận cao hơn mức trung bình từ tài sản của mình.

*   **ROE (Return on Equity):**
    *   Chỉ số này đo lường khả năng sinh lời của một ngân hàng dựa trên vốn chủ sở hữu của cổ đông. ROE cao cho thấy ban lãnh đạo ngân hàng đang sử dụng vốn của cổ đông một cách hiệu quả để tạo ra lợi nhuận.
    *   Một ROE bền vững và tăng trưởng là dấu hiệu của một ngân hàng có quản lý tốt và mô hình kinh doanh hiệu quả.

## So sánh TCB với các ngân hàng khác cùng ngành

```python
import pandas as pd
from vnstock import Listing, Finance

# Get all listed symbols by industry
listing = Listing(source='KBS') # Ensure listing object is initialized
all_symbols_by_industry = listing.symbols_by_industries()

# Find the banking industry (using Vietnamese name 'Ngân hàng' or similar)
bank_industry = all_symbols_by_industry[all_symbols_by_industry['industry_name'].str.contains('Ngân hàng', case=False, na=False)]

# Display some banking symbols found
print("Các mã cổ phiếu ngân hàng được tìm thấy:")
display(bank_industry.head())

# Select a few major banks for comparison, including TCB
# Let's pick a few prominent banks for comparison. Adjust this list as needed.
banks_for_comparison = ['TCB', 'VCB', 'CTG', 'BID', 'MBB', 'ACB']

# Filter to ensure we only process symbols found in the bank_industry list
banks_for_comparison = [bank for bank in banks_for_comparison if bank in bank_industry['symbol'].values]

print(f"\nĐang so sánh các ngân hàng: {', '.join(banks_for_comparison)}")

comparison_data = []

for bank_symbol in banks_for_comparison:
    try:
        finance_obj = Finance(symbol=bank_symbol, source='KBS')
        ratios = finance_obj.ratio(period='year')
        cash_flow_statement = finance_obj.cash_flow(period='year')

        # Extract relevant ratios for the latest year (2025-Năm as per previous outputs)
        pe_ratio = ratios[ratios['item_id'] == 'pe_ratio']['2025-Năm'].values
        pb_ratio = ratios[ratios['item_id'] == 'pb_ratio']['2025-Năm'].values
        nim = ratios[ratios['item_id'] == 'net_interest_margin_nim']['2025-Năm'].values # Net Interest Margin

        # --- Robust ROE extraction (get latest available if 2025-Năm is None) ---
        roe_val = None
        year_cols = [col for col in ratios.columns if '-Năm' in col] # Get all year columns

        # Try the 'averaged' ROE first, then fallback to general ROE
        roe_item_ids_to_check = [
            'roe', # Added this direct item_id for 'Tỷ suất lợi nhuận trên vốn chủ sở hữu bình quân'
            'return_on_equity_roe_averaged',
            'return_on_equity_roe_averaged_4_quarters_latest_year',
            'return_on_equity_roe'
        ]

        for roe_item_id in roe_item_ids_to_check:
            if roe_item_id in ratios['item_id'].values:
                roe_data_row = ratios[ratios['item_id'] == roe_item_id]
                if not roe_data_row.empty:
                    for year_col in sorted(year_cols, reverse=True): # Check from latest year backwards
                        if year_col in roe_data_row.columns and pd.notna(roe_data_row[year_col].iloc[0]):
                            roe_val = roe_data_row[year_col].iloc[0]
                            break # Found latest ROE, break from year loop
                if roe_val is not None: # If found from current item_id, break from item_id loop
                    break

        # --- FCF Calculation for 2025 ---
        fcf_2025 = None
        # Ensure item_id column is string type and strip whitespace for robust comparison
        cash_flow_statement['item_id'] = cash_flow_statement['item_id'].astype(str).str.strip()

        op_cash_flow = cash_flow_statement[cash_flow_statement['item_id'].str.contains('operating_cash_flow', case=False, na=False)]
        capex = cash_flow_statement[cash_flow_statement['item_id'].str.contains('purchase_of_fixed_assets', case=False, na=False)]

        if not op_cash_flow.empty and not capex.empty and '2025-Năm' in op_cash_flow.columns and '2025-Năm' in capex.columns:
            op_cf_val_2025 = op_cash_flow['2025-Năm'].iloc[0]
            capex_val_2025 = capex['2025-Năm'].iloc[0]
            if pd.notna(op_cf_val_2025) and pd.notna(capex_val_2025):
                fcf_2025 = op_cf_val_2025 + capex_val_2025 # CapEx is usually negative, so adding works as subtraction of outflow

        comparison_data.append({
            'Ngân hàng': bank_symbol,
            'P/E (2025)': pe_ratio[0] if len(pe_ratio) > 0 else None,
            'P/B (2025)': pb_ratio[0] if len(pb_ratio) > 0 else None,
            'ROE (2025)': roe_val,
            'NIM (2025)': nim[0] if len(nim) > 0 else None,
            'FCF (2025)': fcf_2025 # Add FCF to the comparison data
        })
    except Exception as e:
        print(f"Could not retrieve data for {bank_symbol}: {e}")

comparison_df = pd.DataFrame(comparison_data)

# Sort by P/E ratio for better comparison, or any other ratio
comparison_df = comparison_df.sort_values(by='P/E (2025)', ascending=True).reset_index(drop=True)

print("\nSo sánh các chỉ số tài chính cơ bản giữa các ngân hàng (Năm 2025):")
display(comparison_df)
```

*   **P/E (Price-to-Earnings):** Cho biết nhà đầu tư sẵn sàng trả bao nhiêu cho mỗi đồng lợi nhuận. P/E thấp hơn có thể gợi ý cổ phiếu bị định giá thấp hoặc triển vọng tăng trưởng không cao bằng các ngân hàng có P/E cao hơn.
*   **P/B (Price-to-Book):** So sánh giá thị trường với giá trị sổ sách của ngân hàng. Chỉ số này đặc biệt hữu ích cho các ngân hàng vì tài sản của họ thường được định giá gần với giá trị sổ sách.

*   **NIM (Net Interest Margin):** Phản ánh hiệu quả kinh doanh cốt lõi của ngân hàng, tức là khả năng tạo ra thu nhập lãi ròng từ các tài sản sinh lãi. NIM cao hơn thường cho thấy hiệu quả hoạt động tốt hơn trong việc quản lý chi phí huy động và lợi suất cho vay.
